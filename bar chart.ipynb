```r
person=c(10,5,5,10,10)
sports=c("Football","Hockey","Cricket","Basketball","Badminton")
```

```r
barplot(person,names.arg=sports,main="Favourite Sports Percentage",xlab="Sports Names",ylab="People",col="darkred")
```

```r
barplot(person,names.arg=sports,main="Favourite Sports Percentage",xlab="Sports Names",ylab="People",horiz=TRUE,col="blue",border="red",density=25)
```
